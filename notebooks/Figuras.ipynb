```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)

x = np.linspace(-10, 10, 100)
ax.plot(x, 0.6*norm.pdf(x*0.3), 'r-', lw=5, alpha=0.6, label='Cq*')
y = (norm.pdf((x-2.5)*2) + norm.pdf((x+1.5)*0.5) + norm.pdf((x-0.5)*1))/3
ax.plot(x, y, 'b-', lw=5, alpha=0.6, label='p*')
ax.vlines(-2.5, 0, 0.6*norm.pdf(-2.5*0.3), 'k', lw=2, alpha=0.9, label='Cq*(-2.5)')

plt.legend()
plt.savefig('../imagenes/rejection_sampling.png')
```

```python
fig, ax = plt.subplots(1, 1)

x = np.linspace(-3, 3, 100)
ax.plot(x, norm.pdf(x), 'b-', lw=5, alpha=0.6, label='p*')
ax.plot(x, norm.pdf(x, scale=1.2), 'r-', lw=5, alpha=0.6, label='q*')
print((norm.pdf(x, scale=1.2)[1:] * (np.exp(x)[1:] + np.exp(x)[:-1])/2 * np.diff(x)).sum())


plt.legend()
plt.savefig('../imagenes/important_sampling.png')
```

```python
fig, ax = plt.subplots(1, 1)

x = norm.rvs(scale=1.2, size=100)

w = norm.pdf(x) / norm.pdf(x, scale=1.2)
w /= w.sum()

print((w * np.exp(x)).sum())
ax.stem(x, w, 'g*', label='Pesos')
ax.set_xlim([-3,3])
plt.legend()
plt.savefig('../imagenes/important_sampling_w.png')
```

```python
fig, ax = plt.subplots(1, 1)

x = np.linspace(-3, 3, 100)
y = (norm.pdf((x-2.5)*5) + norm.pdf((x+1.5)*2) + norm.pdf((x-0.5)*1))/3
ax.plot(x, y, 'b-', lw=5, alpha=0.6, label='p*')

ax.plot(x, 0.6*norm.pdf((x+2)*3.3), 'r-', lw=5, alpha=0.6, label='Q(x1;x0)')

ax.vlines(-2, 0, 0.05, 'k', lw=2, alpha=0.9, label='x0')
ax.vlines(-1.5, 0, 0.05, 'k', lw=1.5, alpha=0.5, label='x1')


plt.legend()
plt.savefig('../imagenes/MH1.png')
```
